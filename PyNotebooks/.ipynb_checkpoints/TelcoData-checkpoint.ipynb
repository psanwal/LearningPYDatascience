```python
# Removing Non-Categorical Columns
TelData_col_list = list(TelData.columns) 
TelData_col_list.remove('customerID')
TelData_col_list.remove('MonthlyCharges')
TelData_col_list.remove('TotalCharges')
TelData_col_list.remove('tenure')

# Removing SeniorCitizen as well because it is already in 0,1 format.
TelData_col_list.remove('SeniorCitizen')

# Checking the unique values and values in each group
for x in TelData_col_list:
    print(x)
    print(TelData[x].value_counts())
    print("Number of Unique Values :{}".format(TelData[x].nunique()))
    print('--------------------')
```

```python
# Checking Corerelated columns if any. We should drop highly co-related columns.
TelData.corr()
```

```python
# As we can see TotalCharges and Tenure column is highly corelated, so we will drop totalCharges
TelData.drop('TotalCharges',axis=1,inplace=True)
```

```python
TelData_new = pd.get_dummies(data=TelData.drop('customerID',axis=1),columns=TelData_col_list, drop_first=True)
TelData_new
```

```python
# Data Preprocessing
```

```python
# Building the model now
```

```python
# importing necessary liberaries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
# Train, Test split
y = TelData_new['Churn_Yes']
x = TelData_new.drop('Churn_Yes',axis=1)
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=101,test_size=0.3)
```

```python
# Scalling Features
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
# Instantiating LogisticRegression variable
lr = LogisticRegression()
lr.fit(x_train,y_train)
lr_predict = lr.predict(x_test)
```

```python
# Printing Classification report
print(classification_report(y_test,lr_predict))
```

```python
# Printing confusion matrix
print(confusion_matrix(y_test,lr_predict))
```

```python
# Score on Training and Test datasets.
# Since Training and test dataset scores are very near this is most probably a underfit model.
# We can try increasing C, which will increase the complexity of the model.
# Or we can decrease C, which will regularize the model and models performance in training set will decrease but modle will be more generalized.
print('Training Dataset score : {:.3f}'.format(lr.score(x_train,y_train)))
print('Test Dataset score : {:.3f}'.format(lr.score(x_test,y_test)))
```

```python
# Building Same model for different values of c [.001,.01,1,10,100,1000]
# using L1 and L2 Regularization
for r in ['l1','l2']:
    print('---------------For Penalty = {}---------'.format(r))
    for c in [.001,.01,1,10,100,1000]:
        print('---------------For C= {}---------'.format(c))
        lr = LogisticRegression(penalty=r,C=c)
        lr.fit(x_train,y_train)
        lr_predict = lr.predict(x_test)
        print('Training Dataset score : c={},penalty={}, {:.3f}'.format(c,r,lr.score(x_train,y_train)))
        print('Test Dataset score : c={},penalty={}, {:.3f}'.format(c,r,lr.score(x_test,y_test)))
        print(lr.coef_)
        
    
```
